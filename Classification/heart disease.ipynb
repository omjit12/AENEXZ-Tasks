```python
plt.figure(figsize=(10, 6))
sns.countplot(x='chest pain type', hue='target', data=df, palette='viridis')
plt.title('Heart Disease Frequency for Chest Pain Type', fontsize=15)
plt.xlabel('Chest Pain Type (1-4)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(["No Disease", "Disease"])
plt.show()
```

"Asymptomatic" (Type 4) type actually show higher disease frequency than "Typical" pain. This is a critical finding for the model—it suggests that heart disease often presents in ways that don't feel like "classic" heart pain.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select only numerical columns for the box plots
numerical_columns = ['age', 'resting bps', 'cholesterol', 'max heart rate', 'oldpeak']

# 2. Create the figure and subplots
plt.figure(figsize=(16, 10))

# 3. Loop through columns to create individual box plots
for i, col in enumerate(numerical_columns):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(data=df, y=col, color='#5DADE2', width=0.5)
    plt.title(f'Box Plot of {col}', fontsize=14)
    plt.ylabel('Value Range')

# 4. Final layout adjustments
plt.tight_layout()
plt.suptitle('Outlier Detection in Clinical Features', fontsize=18, y=1.05)
plt.show()
```

```python
Q1 = df['cholesterol'].quantile(0.25)
Q3 = df['cholesterol'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['cholesterol'] < lower) | (df['cholesterol'] > upper)].index
```

```python
df = df.iloc[~df.index.isin(outliers)]
```

# Logistic Regression

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
acc_lr = accuracy_score(y_test, lr.predict(X_test))
```

# Decision Tree


```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
acc_dt = accuracy_score(y_test, dt.predict(X_test))
```

# KNN

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
acc_knn = accuracy_score(y_test, knn.predict(X_test))
```
